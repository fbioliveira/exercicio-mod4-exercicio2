## Exercício 1: Vestibular

Considere que a os dados gerados na célula abaixo contêm o número de acertos de 100 alunos em um vestibular para um curso de exatas, divididas pelos respectivos assuntos. Considere que cada assunto possui um número de questões conforme a tabela abaixo:

| assunto | número de questões |
|:---:|:---:|
| Matemática | 24 |
| Português | 18 |
| Geografia | 8 |
| Inglês | 8 |
| História | 8 |
| Física | 12 |
| Química | 12 |

Usando os comandos de operações com DataFrames que você aprendeu na Aula 03, calcule:

1. (operações com escalar) Calcule o percentual de acerto dos alunos por assunto.  
2. (operações entre *DataFrames) Calcule o total de acertos de cada aluno.  
3. Calcule o porcentual geral de cada aluno.  
4. Suponha que a nota de corte para a segunda fase seja 45. Quantos alunos tiveram nota maior que 45?  

````python
import pandas as pd
import numpy as np

np.random.seed(42)
df_mat = pd.DataFrame(np.random.randint(24, size=(100, 1)), columns=['Qt_acertos'])

df_por = pd.DataFrame(np.random.randint(18, size=(100, 1)), columns=['Qt_acertos'])

df_geo = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_ing = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_his = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_fis = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos'])

df_qui = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos'])

````

## Observação:

O código acima não dá a possibilidade de alcançar a nota máxima, no caso eu poderia puxar os valores pelo len() da variavel np.unique(), porém nesse caso a disciplina de português não iria vir correta, pois ela não alcancou o valor máximo devido o 'seed(42)', que deveria ser 17.

````python
valores_mat = df_mat['Qt_acertos'].values
valores_unicos_mat = np.unique(valores_mat)

valores_por = df_por['Qt_acertos'].values
valores_unicos_por = np.unique(valores_por)

valores_geo = df_geo['Qt_acertos'].values
valores_unicos_geo = np.unique(valores_geo)

valores_ing = df_ing['Qt_acertos'].values
valores_unicos_ing = np.unique(valores_ing)

valores_his = df_his['Qt_acertos'].values
valores_unicos_his = np.unique(valores_his)

valores_fis = df_fis['Qt_acertos'].values
valores_unicos_fis = np.unique(valores_fis)

valores_qui = df_qui['Qt_acertos'].values
valores_unicos_qui = np.unique(valores_qui)

print(f'Matemática: {valores_unicos_mat}')
print(f'Português: {valores_unicos_por}')
print(f'Geografia: {valores_unicos_geo}')
print(f'Inglês: {valores_unicos_ing}')
print(f'História: {valores_unicos_his}')
print(f'Física: {valores_unicos_fis}')
print(f'Química: {valores_unicos_qui}')

print(f'\nA quantidade de elementos em português é {len(valores_unicos_por)}, sendo que o valor máximo de nota que o aluno pode alcançar é {np.amax(valores_unicos_por)}')
````

````python
# 1) Seu código aqui

pd.options.display.float_format = "{:,.2f}".format
df_mat.rename(columns={'Qt_acertos':'mat'}, inplace=True)
df_por.rename(columns={'Qt_acertos':'por'}, inplace=True)
df_geo.rename(columns={'Qt_acertos':'geo'}, inplace=True)
df_ing.rename(columns={'Qt_acertos':'ing'}, inplace=True)
df_his.rename(columns={'Qt_acertos':'his'}, inplace=True)
df_fis.rename(columns={'Qt_acertos':'fis'}, inplace=True)
df_qui.rename(columns={'Qt_acertos':'qui'}, inplace=True)
df_mat['%'] = (df_mat['mat']/df_mat['mat'].nunique())*100
df_por['%'] = (df_por['por']/df_por['por'].nunique())*100
df_geo['%'] = (df_geo['geo']/df_geo['geo'].nunique())*100
df_ing['%'] = (df_ing['ing']/df_ing['ing'].nunique())*100
df_his['%'] = (df_his['his']/df_his['his'].nunique())*100
df_fis['%'] = (df_fis['fis']/df_fis['fis'].nunique())*100
df_qui['%'] = (df_qui['qui']/df_qui['qui'].nunique())*100
df_completo = pd.concat([df_mat,df_por,df_geo,df_ing,df_his,df_fis,df_qui], axis=1)
df_completo
````

````python
# 2) Seu código aqui
df_completo['total_acertos'] = df_completo['mat'] + df_completo['por'] + df_completo['geo'] + df_completo['ing'] + df_completo['his'] + df_completo['fis'] + df_completo['qui']
df_completo
````

````python
# 3) Seu código aqui
total_questoes = df_completo['mat'].nunique() + df_completo['por'].nunique() + df_completo['geo'].nunique() + df_completo['ing'].nunique() + df_completo['his'].nunique() + df_completo['fis'].nunique() + df_completo['qui'].nunique()
print(total_questoes)
df_completo['% geral de acertos'] = df_completo['total_acertos'] / total_questoes
df_completo
````

````python
# 4) Seu código aqui

df_alunos_nota_corte = df_completo[df_completo['total_acertos']>45]
print(f'Qtd. de alunos que passaram na nota de corte: {len(df_alunos_nota_corte)}')
df_alunos_nota_corte
````

4. Monte um *DataFrame* com os alunos que passaram para a segunda fase. Repare que estes alunos não podem ter sido desqualificados.

````python
# seu código aqui
desqualificados_mat = df_completo[df_completo['mat'] == 0]
desqualificados_fis = df_completo[df_completo['fis'] == 0]
desqualificados_qui = df_completo[df_completo['qui'] == 0]

desqualificados_geral = pd.concat([desqualificados_mat, desqualificados_fis, desqualificados_qui])
desqualificados_geral.drop_duplicates(inplace=True, keep='first')

print(f'A quantidade de alunos desqualificados é: {len(desqualificados_geral)}')
print(f'A média dos alunos em História é: {desqualificados_geral["his"].mean():.2f}')
print(f'A média dos alunos em Geografia é: {desqualificados_geral["geo"].mean():.2f}')

print(f'\nAlunos que passaram para a segunda fase: ')
aprovados_segunda_fase = df_completo[(df_completo.mat > 0) & (df_completo.fis > 0) & (df_completo.qui > 0)]
print(f'\nQuantos zeraram em Matematica?: {aprovados_segunda_fase.mat.isnull().sum()}')
print(f'Quantos zeraram em Física?: {aprovados_segunda_fase.fis.isnull().sum()}')
print(f'Quantos zeraram em Química?: {aprovados_segunda_fase.qui.isnull().sum()}')

aprovados_segunda_fase



````
